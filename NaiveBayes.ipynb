# Import libraries

```python
import numpy as np
import pandas as pd
from sklearn import metrics
```

# Load Data

```python
train_data = pd.read_csv('fashion-mnist_train.csv')
train_data.head()
```

```python
test_data = pd.read_csv('fashion-mnist_test.csv')
test_data.head()
```

# Data Selection

```python
X_train = train_data.iloc[:, 1:].values
Y_train = train_data.iloc[:, 0].values
X_test = test_data.iloc[:, 1:].values
Y_test = test_data.iloc[:, 0].values
```

# Modelling
For Naive Bayes we model differnt algorithms for multiclassification and then take the model which gives the best result

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, Y_train)
predictions = classifier.predict(X_test)
print('Accuracy is :', metrics.accuracy_score(Y_test, predictions))
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(X_train, Y_train)
predictions = classifier.predict(X_test)
print('Accuracy is :', metrics.accuracy_score(Y_test, predictions))
```

```python
from sklearn.naive_bayes import ComplementNB
classifier = ComplementNB()
classifier.fit(X_train, Y_train)
predictions = classifier.predict(X_test)
print('Accuracy is :', metrics.accuracy_score(Y_test, predictions))
```

```python
from sklearn.naive_bayes import BernoulliNB
classifier = BernoulliNB()
classifier.fit(X_train, Y_train)
predictions = classifier.predict(X_test)
print('Accuracy is :', metrics.accuracy_score(Y_test, predictions))
```

From the above models, we get that Bernoulli Naive Bayes gives the best result

```python
from sklearn.metrics import confusion_matrix, classification_report
cfm = pd.DataFrame(confusion_matrix(Y_test, predictions))
cfm.columns= range(10)
cfm.index = range(10)
cfm
```
